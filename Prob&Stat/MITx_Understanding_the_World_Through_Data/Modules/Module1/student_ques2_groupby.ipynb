## Average Yearly Temperature - Part 1

Create a new DataFrame called, paris, that only has data for Paris. You'll need to filter the original data for rows whose city matches "Paris". Then use the paris DataFrame to create a third DataFrame, called parismean, that specifies the average temperature per year in Paris. To do this, you'll need to use the groupby function. Provide the average temperature in Paris for each of the years.

```python
paris = df.loc[df['City'] == 'Paris']
parismean = paris.groupby('Year').mean()
parismean.sort_values('AvgTemperature')
```

## Average Yearly Temperature - Part 2
Now that we have the parismean DataFrame, we can find the year in which the average temperature was the lowest. To do this, first find the minimum average temperature and save it as a variable called minparismean. Then, locate the row in the parismean DataFrame whose temperature matches minparismean.

Which year had the minimum average temperature in Paris?

```python
minparismean = parismean['AvgTemperature'].min()
parismean.loc[parismean['AvgTemperature'] == minparismean] 
```

## Number of Days above 80 - Part 1

Create a new DataFrame called `hightemp`, that only has data where the temperature is **above 80 degrees**. You'll need to filter the original data for rows whose temperature is greater than 80. Then use the `hightemp` DataFrame to create a third DataFrame called `cities` that specifies the number of days above 80 per city. To do this, you'll need to use the groupby function. Provide the count for the number of days above 80 degrees for the following cities.

```python
hightemp = df.loc[df['AvgTemperature'] > 80] 
cities = hightemp.groupby('City').count()
print(cities.loc[cities['Day'] == 5441])
print(cities.loc[cities['Day'] == 136])
```

## Number of Days above 80 - Part 2

Now that we have the `cities` DataFrame, we can find the largest of the count values from this DataFrame. To do this, first find the **maximum** count and save it as a variable called `highestcount`. Then, locate the row in the cities DataFrame whose count matches `highestcount`.
<br><br>
Which city had the largest number of days over 80 degrees?

```python
# Your code here
highestcount = cities['Day'].max()
cities.loc[cities['Day'] == highestcount]
```
