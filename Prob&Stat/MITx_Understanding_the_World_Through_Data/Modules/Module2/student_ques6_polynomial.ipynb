# Calculating Gravity

```python
import pandas as pd
import sklearn.linear_model as lm
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv("./Data/planets.csv")
df
```

```python
# your code here to find the quadratic relationship between time and planet_1
time = df[['time']]
time_trans = PolynomialFeatures(degree=2)
time_trans = time_trans.fit_transform(time)
p_1 = df['planet_1']
lin_reg = lm.LinearRegression()
lin_reg.fit(time_trans, p_1)
print('Coefficients:', lin_reg.coef_)
print('Intercept:', lin_reg.intercept_)
```
